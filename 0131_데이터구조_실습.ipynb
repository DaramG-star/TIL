## 1449. 정답

```python
# 아래 함수를 수정하시오.
def intersection_sets(set1, set2):
    intersection = set1 & set2  # 교집합 구하기
    if intersection:
        return len(intersection), intersection
    else:
        print("공통 요소가 없습니다.")
        return 0, set()


result = intersection_sets({1, 2, 3}, {3, 4, 5})
print(result)  # (1, {3})

result = intersection_sets({1, 2}, {3, 4})
print(result)  # (0, set())
# 출력: 공통 요소가 없습니다
```

## 1451. 차집합 Set.difference() Lv5

image.png

```python
# 주어진 코드
def ordered_difference_sets(set1, set2):
    difference1 = set1 - set2
    difference2 = set2 - set1
    if len(difference1) > len(difference2):
      return difference2, difference1
    elif len(difference2) > len(difference1):
      return difference1, difference2
    else :
      return difference1, difference2

# 예시 실행
result = ordered_difference_sets({1, 2, 3, 4}, {3, 4, 5, 6})
print("결과:", result)  # 출력: ({1, 2}, {5, 6})

result = ordered_difference_sets({1, 2, 3, 4}, {1, 2, 3})
print("결과:", result)  # 출력: (set(), {4})
```
