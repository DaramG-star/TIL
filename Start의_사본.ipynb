```python
# 2 10  

tar = 10010101
decimal_value = 0
power = 0

while tar > 0:
    last_digit = tar%10
    decimal_value += last_digit * (2 ** power)
    tar //= 10
    power += 1

print(decimal_value)
```

```python
# 16 10  

hex_num = "1A3"  # 16 
decimal_value = 0  #  10 
power = 0  # 16  ()

for i in range(len(hex_num) - 1, -1, -1):  # (1 )  
    digit = hex_num[i]  #    ()

    # 16 A~F 10~15 
    if '0' <= digit <= '9':
        value = ord(digit) - ord('0')  #  '0' ASCII (48)   
    else:
        value = ord(digit) - ord('A') + 10  #  'A' ASCII (65) 'A' 10 

    decimal_value += value * (16 ** power)  # 16   
    power += 1  #  

print(decimal_value)  # 10 
```

```python
# 16 2 

hex_to_bin = {
    '0': '0000', '1': '0001', '2': '0010', '3': '0011',
    '4': '0100', '5': '0101', '6': '0110', '7': '0111',
    '8': '1000', '9': '1001', 'A': '1010', 'B': '1011',
    'C': '1100', 'D': '1101', 'E': '1110', 'F': '1111'
}

T = int(input())
for t in range(1, T+1):
    N, hexx = input().split()

    ans = ''.join(hex_to_bin[h] for h in hexx)

    print(f"#{t} {ans}")
```

```python
M = 31
N = 5
def Test():
    tar = M
    for i in range(N):
        if tar & 0x1 == 0:
            return False
        tar = tar >> 1
    return  True

print(Test())
```
