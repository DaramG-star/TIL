```python
food_list = [
    {
        '종류': '한식',
        '이름': '잡채'
    },
    {
        '종류': '채소',
        '이름': '토마토'
    },
    {
        '종류': '중식',
        '이름': '자장면'
    },
]

# 아래에 코드를 작성하시오.
for food in food_list :
    if food['이름'] == '토마토':
        food['종류'] = '과일'
        print(f"{food['이름']}은/는 {food['종류']}이다.")
    elif food['이름'] == '자장면':
        print('자장면엔 고춧가루지')
        print(f"{food['이름']}은/는 {food['종류']}이다.")
    else :
        print(f"{food['이름']}은/는 {food['종류']}이다.")

print(food_list)
```

```python
import requests

# 블랙리스트 정의
black_list = [
    'Hoeger LLC',
    'Keebler LLC',
    'Yost and Sons',
    'Johns Group',
    'Romaguera-Crona',
]

# 사용자 이름을 저장할 리스트
dummy_data = []

# 1번부터 10번까지 사용자 데이터를 요청
for i in range(1, 11):  # ID가 1부터 10까지
    API_URL = f'https://jsonplaceholder.typicode.com/users/{i}'
    response = requests.get(API_URL)
    parsed_data = response.json()

    name = parsed_data['name']
    lat = float(parsed_data['address']['geo']['lat'])
    lng = float(parsed_data['address']['geo']['lng'])
    company_name = parsed_data['company']['name']

    # 위도와 경도가 조건에 맞는 경우에만 추가
    if (lat < 80 and lat > -80) and (lng < 80 and lng > -80):
        dummy_data.append({
            'company': company_name,
            'lat': lat,
            'lng': lng,
            'name': name,
        })

# 2차 필터링: 블랙리스트 검열 함수
def censorship(filtered_users):
    censored_user_list = {}
    for user in filtered_users:
        company = user['company']
        name = user['name']
        # 블랙리스트에 포함된 경우 메시지 출력
        if company in black_list:
            print(f"{company} 소속의 {name}은/는 등록할 수 없습니다.")
        else:
            print(f"이상 없습니다.")
            if company not in censored_user_list:
                censored_user_list[company] = []
            censored_user_list[company].append(name)
    return censored_user_list

# 검열 실행
censored_user_list = censorship(dummy_data)

# 결과 출력
print(censored_user_list)
```

```python
blood_types = ['A-', 'A+', 'B-', 'B+', 'O-', 'O+', 'AB-', 'AB+']
black_list = [
    'Jenkins-Garcia',
    'Stephens Group',
    'White, Andrade and Howard',
    'Warren-Stewart',
]

# 유효성 검사 함수
def is_validation(user):
    global num_false  # 전역 변수로 False 개수를 추적
    num_false = 0

    # 유효성 검사 조건
    if user['blood_group'] not in blood_types:
        num_false += 1
        return False
    if user['company'] in black_list:
        print(f"Blocked: {user['company']} 소속의 {user['name']}은/는 등록할 수 없습니다.")
        return False
    if '@' not in user['mail']:
        num_false += 1
        return False
    if len(user['name']) < 2 or len(user['name']) > 30:
        num_false += 1
        return False
    if len(user['website']) == 0:
        num_false += 1
        return False

    return True  # 모든 조건을 통과하면 True 반환

# 사용자 데이터를 처리하고 유효한 사용자 리스트 생성
def create_user(user_data):
    global num_false  # 전역 변수 사용
    num_false = 0
    valid_users = []

    for user in user_data:
        if is_validation(user):
            valid_users.append(user)

    print(f"유효하지 않은 데이터 개수: {num_false}")
    return valid_users

# 사용자 데이터 처리 실행
valid_user_list = create_user(user_data)

# 유효한 사용자 리스트 출력
print("\n최종 유효 사용자 리스트:")
for user in valid_user_list:
    print(f"{user['name']} (Company: {user['company']}, Blood Group: {user['blood_group']})")
```

```python
# len 구하기
def get_len(arr):
    cnt = 0
    for num in arr:
        cnt += 1
    return cnt

# max
def get_max(arr):
    max_v = arr[0]   # -987654321
    for num in arr:
        if max_v < num:
            max_v = num
    return max_v

# min
def get_min(arr):
    min_v = arr[0]  # 987654321
    for num in arr:
        if min_v > num:
            min_v = num
    return min_v

# sum
def get_sum(arr):
    sum_v = 0
    for num in arr:
        sum_v += num
    return sum_v

arr = [3, 2, 1, 4, 5, 32]
print(get_len(arr))
print(get_max(arr))
print(get_min(arr))
print(get_sum(arr))

# arr 변수의 평균을 구하시오.
sum_v = get_sum(arr)
N = get_len(arr)
avg = sum_v / N
print(round(avg, 2))
print(f'{avg:.0f}')
print(int(avg))

# 중복제거하기
def get_set(arr):
    result = []
    for num in arr:
       if num not in result:
           result.append(num)
    return result

arr = [1, 3, 4, 3, 5, 3]
print(list(set(arr)))
print(get_set(arr))


# 추후에는 max_idx, min_idx 를 구할 줄 알아야 한다.
```
